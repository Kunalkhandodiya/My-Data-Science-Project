```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
print(iris.target_names)
print(iris.feature_names)
```

```python
import pandas as pd
data=pd.DataFrame({'sepal length':iris.data[:,0],
              'sepal width':iris.data[:,1],
              'petal length':iris.data[:,2],
             'petal width':iris.data[:,3],
             'species':iris.target })
```

```python
data.head()
```

```python
x=data[['sepal length','sepal width','petal length','petal width','species']]# features
y=data['species'] # lables
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3) # 70% gives Training set ,30% gives test 
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
# import Random classifier
from sklearn.ensemble import  RandomForestClassifier
```

```python
c=RandomForestClassifier(n_estimators=100,criterion='gini')
```

```python
c.fit(x_train,y_train)
```

```python
c.predict(x_test)
```

```python
x_test # (0: setosa , 1: versicolor, 2: virginica)
```

```python
# check the scrore of this model
c.score(x_test,y_test)
```
